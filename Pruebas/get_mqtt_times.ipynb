```python
import pandas as pd
```

```python
file_arrive_time = "TimeMessageArrivedToMCU.csv"
df_arrive = pd.read_csv(file_arrive_time)
df_arrive.head(5)
```

```python
file_depart_time = "TimeMessageDepartFromApp.csv"
df_depart = pd.read_csv(file_depart_time)
df_depart.head(5)
```

```python
import datetime
from datetime import timezone

df_time_analysis = pd.DataFrame()
df_time_analysis['time_message_arrive_to_mcu'] =  pd.to_datetime(df_arrive['data'].str.slice(stop=19) + '.' + df_arrive['ms'].str.slice(stop=4))
#df_time_analysis['time_message_arrive_to_mcu'] = df_time_analysis['time_message_arrive_to_mcu'].dt.tz_localize('UTC').dt.tz_convert('Etc/GMT+4')
df_time_analysis = df_time_analysis.iloc[::2]
df_time_analysis.reset_index(inplace=True,drop=True)
df_time_analysis.head(5)
```

```python
import numpy as np
df_time_analysis['time_message_depart_from_app'] = pd.to_datetime(df_depart['data'],unit='ms',origin='unix')
df_time_analysis['time_message_depart_from_app'] = df_time_analysis['time_message_depart_from_app'].dt.tz_localize('UTC').dt.tz_convert('Etc/GMT+4').dt.tz_localize(None)
df_time_analysis['delta_time'] = df_time_analysis['time_message_arrive_to_mcu'] - df_time_analysis['time_message_depart_from_app']
df_time_analysis['delta_time_ms'] = (df_time_analysis['time_message_arrive_to_mcu'] - df_time_analysis['time_message_depart_from_app']).astype(np.int64) // 10**6
df_time_analysis
```

```python
analysis_file_name = "TimeAnalysis.xlsx"
df_time_analysis.to_excel(analysis_file_name,sheet_name="time_data")
```
